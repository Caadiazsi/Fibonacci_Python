```python
from __future__ import unicode_literals
import random;
from time import time;
import math
import numpy as np
import pylab as pl    
```

```python
def Fibonacci (N):
    one = 0
    two = 1
    Help = 0
    initial_time = time();
    if N==2:
        Help = 1;
    else:
        for x in range (2, N):
            Help = one + two
            one = two
            two = Help
    return time()-initial_time
```

```python
print("Ingrese N")
N=int(input())
```

```python
Times = []
for x in range (1,N):
    Times.append(Fibonacci(x))
Numbers = np.arange(1,N,1)
pl.xlabel("N")
pl.ylabel("Time")
pl.plot(Numbers,Times)
```
